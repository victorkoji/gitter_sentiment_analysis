```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

#Nome da pasta tema
pasta_tema = "BOT"

#Nome do arquivo
nome_arquivo = "basic_bot"

usuario = "Sidewinder"

#Caminho do arquivo
path_arquivo = f"./ChatRooms/{pasta_tema}/{nome_arquivo}/{nome_arquivo}"

#Visualizando os dados:
df = pd.read_csv(fr"{path_arquivo}_threads_classificado.csv",  encoding='utf-8')

usuarioMessages = df[df['Username'] == usuario]
usuarioMessages
```

```python
#Quantidade de valores classificados: positivos, neutros e negativos.
positive = len(df[df['Classificação'] == 'Positive'])
neutral = len(df[df['Classificação'] == 'Neutral'])
negative = len(df[df['Classificação'] == 'Negative'])

#Define a quantidade de coluna e os seus valores
x = np.arange(3)
values = [positive, neutral, negative]

#Monta o gráfico
fig, ax = plt.subplots()
fig.patch.set_facecolor('white')
plt.bar(x, values)
plt.xticks(x, ('Positive', 'Neutral', 'Negative'))

#Salva a imagem
plt.savefig(f"{path_arquivo}_threads_grafico.png", transparent=False)

#Mostra a imagem
plt.show()
```
